# 13. 이미지 변형(이진화) - threshold
- 특정값을 기준( 임계값 )으로 흑백으로 나누는 바이너리 이미지 변환 방법

```python
import cv2
img = cv2.imread("book.jpg",cv2.IMREAD_GRAYSCALE)

# 이미지, 임계값 , 클때 적용, 바이너리
ret,bineary = cv2.threshold(img,127,255,cv2.THRESH_BINARY)

cv2.imshow("img",img)
cv2.imshow("bineary",bineary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### trackbar( 값 변화에 따른 변형 확인 ) 
-일정 범위 내의 값을 변경할 때 사용하며, 적절한 임곗값을 찾거나 변경하기 위해 사용

```python
import cv2

def empty(pos):
    print(pos)
    pass

img = cv2.imread("threshold.png",cv2.IMREAD_GRAYSCALE)

name = "trackber"

cv2.namedWindow(name)

cv2.createTrackbar("threshold",name,127,255,empty) # bar이름, 창의 이름, 초깃값, 최대값, 이벤트처리

while True:
    thresh = cv2.getTrackbarPos('threshold',name) # bar이름, 창이름
    ret,bineary = cv2.threshold(img,thresh,255,cv2.THRESH_BINARY)

    if not ret:
        break
    cv2.imshow(name,bineary)
    if cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()
```

### 임계값의 경계 
- 검은색 : 0 
- 진한 회색 : 127 
- 밝은 회색 : 195
- 흰색 : 255

```python
import cv2


img = cv2.imread("threshold.png",cv2.IMREAD_GRAYSCALE)

ret,bineary1 = cv2.threshold(img,0,255,cv2.THRESH_BINARY)       #w진환 회색, 밝은 회색, 흰색 처리 
ret,bineary2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)     # 밝은 회색, 흰색 
ret,bineary3 = cv2.threshold(img,195,255,cv2.THRESH_BINARY)     # 흰색 


cv2.imshow("img",img)
cv2.imshow("img1",bineary1)
cv2.imshow("img2",bineary2)
cv2.imshow("img3",bineary3)

cv2.waitKey(0)
    
cv2.destroyAllWindows()
```

### Adaptive Threshold
-> 영역을 세분화 하여 임계치를 적용하는 방법( 햇빛, 조명, 명암 등등의 이미지 )

```python
import cv2

def empty(pos):
    print(pos)
    pass

img = cv2.imread("book.jpg",cv2.IMREAD_GRAYSCALE)

name = "trackber"

cv2.namedWindow(name)

cv2.createTrackbar("block_size",name,25,100,empty) # 홀수만 가능 , 1보다는 큰값 
cv2.createTrackbar("c",name,3,10,empty) # 일반적으로 양수의 값을 사용 

while True:
    block_size = cv2.getTrackbarPos('block_size',name) # bar이름, 창이름
    c =cv2.getTrackbarPos("c",name) 
    
    if block_size<=1:
        block_size=3
    if block_size%2==0:
        block_size+=1
    
    bineary = cv2.adaptiveThreshold(img,255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,blockSize=block_size,C=c)

    
    cv2.imshow(name,bineary)
    if cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()
```

## 오츠 알고리즘 
- 최적의 임계치를 자동으로 찾는 알고리즘 
- 이미지 형태가 bimoadi image에 사용하기 적합 

```python
import cv2 
img = cv2.imread("book.jpg",cv2.IMREAD_GRAYSCALE)

ret,bineary = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,ostu = cv2.threshold(img,-1,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU) 
print("ostu",ret)


cv2.imshow("img",img)
cv2.imshow("bineary",bineary)
cv2.imshow("ostu",ostu)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
